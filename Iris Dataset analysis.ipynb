```python
import pandas as pd
```

```python
data = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
read_data = pd.read_csv(data)
print(read_data)
```

```python
#calculating the mean petal length for the setosa species
filter_setosa = read_data[read_data['species'] == 'setosa']
print(filter_setosa)

petal_mean = filter_setosa['petal_length'].mean()
print('The petal mean of setosa species is:',petal_mean)
```

```python
# Calculate petal area for all species
# Getting petal area = petal length * petal width
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

read_data['petal_area'] = read_data['petal_length'] * read_data['petal_width']
read_data

```

```python
# Calculating mean petal area for all species 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
mean_petal_area = read_data.groupby('species')['petal_area'].mean()
data_mean_petal_area = pd.DataFrame({'mean_petal_area': mean_petal_area})

print(data_mean_petal_area)
```

```python
# Calculating mean delta between petal width and sepal width for virginica species
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
filter_virginica = read_data[read_data['species'] == 'virginica'].copy()

# differences of sepal width and petal width 
filter_virginica.loc[:, 'delta'] = filter_virginica['sepal_width'] - filter_virginica['petal_width']

print(filter_virginica)

mean_delta = filter_virginica['delta'].mean()
print(mean_delta)
```
